```python
import pandas as pd # Kutuphaneyi yukledik.

```

```python
"""csv : Comma Separated Values/ Virgül ile ayrılmış değerler."""

titanic = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
# Dosyayı ilgili linkten çekiyoruz. Dosyayı pc ye indirip, adresini aynı yere yapıştırarak ta aynı işlemi yapabiliriz
```

```python
titanic.head() # .head() ilk 5 satırı ekrana getirir. (10) yaparsan ilk 10 u getirir.
```

```python
titanic.info() # .info() veris etindeki kolanları, her kolonun ver tiplerini, eksik değer sayısını gösterir.
```

```python
titanic.isnull().sum() # .isnull() her bir hücrenin eksik (null) olup olmadığını kontrol eder.
# .sum() bu eksik değerlerin toplam syısını hesaplarız.
```

```python
titanic['Age']=titanic['Age'].fillna(titanic['Age'].mean())
# .fillna() ile, titanic klasöründe Age bölümündeki boşlukları, Age'in ortalaması ile dolduruyoruz.
```

```python
titanic=titanic.dropna(subset=['Embarked']) # .dopna() ile, Embarked kolonundaki eksik değerleri siliyoruz.
```

```python
titanic.isnull().sum() # Güncellenmiş veriyi kontrol ediyoruz.
```

```python
import seaborn as sns # Veri görselleştirme kütüphanesi.
import matplotlib.pyplot as plt    # Daha genel bir görselleştirme kütüphanesi,çizim yapabilmek için güçlüdür.
```

```python
sns.countplot(x="Sex", hue="Survived", data=titanic) # Belirli bir kategorik değişkenin frakanslarını görselleştirmek için kullanrılır.
"""
x=Sex, e göre
hue=Survived, durumununu renklere göre ayırır.
"""
plt.title("Cinsiyete göre hayatta kalma durumu") # Grafiklere başlık ekleyebiliriz.
plt.show()

sns.countplot(x="Pclass", hue="Survived", data=titanic)
plt.title("Sosyal sınıfa göre hayatta kalma durumu")
plt.show()
```
